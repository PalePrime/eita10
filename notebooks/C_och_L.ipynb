```python
from lcapy import *
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
cct1 = Circuit("""
    Vs  1  0 {rampstep((t-0.005)*100) - rampstep((t-0.02)*50)}; down, l=$v(t)$
    W0  1 _1; right, i>^=$i(t)$
    C  _1 _0 10e-6; down, l^={$C=10 \mu \mathrm{F}$}
    W1 _0  0; left
    ; draw_nodes=false, label_nodes=false
    """)
cct1.draw()
```

```python
voltage = cct1[1].v
current = cct1.C.i
power   = voltage * current
energy  = cct1.C.C * (voltage ** 2) / 2

vt = np.linspace(0, 0.05, 200)
```

```python
%matplotlib widget

vt = np.linspace(0, 0.05, 200)
ticks_milli = mpl.ticker.FuncFormatter(lambda x, pos: '{0:g}'.format(x * 1000))
ticks_micro = mpl.ticker.FuncFormatter(lambda x, pos: '{0:g}'.format(x * 1e6))

fig, axs = plt.subplots(2, 2, figsize=(8, 6), constrained_layout=True, sharex=True, label="Capacitor stored energy")

for a in axs.flatten():
    a.xaxis.set_major_formatter(ticks_milli)
    a.axvspan(5e-3, 15e-3, facecolor = '0.8')
    
axs[0, 1].yaxis.set_major_formatter(ticks_milli)
axs[1, 0].yaxis.set_major_formatter(ticks_milli)
axs[1, 1].yaxis.set_major_formatter(ticks_micro)

voltage.plot(vt, axes=axs[0, 0], color='orange', xlabel='',          ylabel='Capacitor voltage (V)')
current.plot(vt, axes=axs[0, 1], color='red',    xlabel='',          ylabel='Charging current (mA)')
power  .plot(vt, axes=axs[1, 0], color='blue',   xlabel='Time (ms)', ylabel='Charging power (mW)')
energy .plot(vt, axes=axs[1, 1], color='green',  xlabel='Time (ms)', ylabel='Stored energy ($\mu$J)')

display()
```

```python
cct2 = Circuit("""
    Is  1  0 {0.1*rampstep((t-0.005)*100) - 0.1*rampstep((t-0.02)*50)}; down, l=$i(t)$
    W0  1 _1; right
    L  _1 _0 100e-3; down, l^={$L=100 \mathrm{mH}$}, v<=$v(t)$
    W1 _0  0; left
    ; draw_nodes=false, label_nodes=false
    """)
cct2.draw()
```

```python
current2 = cct2.Is.i
voltage2 = cct2[1].v
power2   = voltage2 * current2
energy2  = cct2.L.L * (current2 ** 2) / 2

vt2 = np.linspace(0, 0.05, 200)
```

```python
fig2, axs2 = plt.subplots(2, 2, figsize=(8, 6), constrained_layout=True, sharex=True, label="Inductor stored energy")

for a in axs2.flatten():
    a.xaxis.set_major_formatter(ticks_milli)
    a.axvspan(5e-3, 15e-3, facecolor = '0.8')
    
axs2[0, 0].yaxis.set_major_formatter(ticks_milli)
axs2[1, 0].yaxis.set_major_formatter(ticks_milli)
axs2[1, 1].yaxis.set_major_formatter(ticks_micro)

current2.plot(vt2, axes=axs2[0, 0], color='orange', xlabel='',          ylabel='Charging current (mA)')
voltage2.plot(vt2, axes=axs2[0, 1], color='red',    xlabel='',          ylabel='Inductor voltage (V)')
power2  .plot(vt2, axes=axs2[1, 0], color='blue',   xlabel='Time (ms)', ylabel='Charging power (mW)')
energy2 .plot(vt2, axes=axs2[1, 1], color='green',  xlabel='Time (ms)', ylabel='Stored energy ($\mu$J)')

display()
```
