```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
```

```python
sos = signal.ellip(8, 1, 60, 0.12, fs=1, output='sos')
```

```python
sos
```

```python
f, a = signal.sosfreqz(sos, fs=1.0)
```

```python
%matplotlib widget
p = plt.plot(f, 20 * np.log(abs(a)))
```
